```python
from os import listdir
from os.path import isfile, join
from matplotlib import pyplot as plt

txt_files = [f for f in listdir("text_from_img") if isfile(join("text_from_img", f))]

def get_files_from_text(phrase):
    phrase = phrase.split()
    file = ""
    out = []
    for i in txt_files:
        with open(r"text_from_img/"+i, "r", encoding="utf-8") as f:
            file = f.read()
        
        for word in phrase:
            if word in file:
                out.append(i[:-4])
    
    return out

def print_img(str):
    img = plt.imread('img_crop/'+str+'.jpg')
    plt.figure(figsize=(17, 17))
    plt.imshow(img)
    plt.show()


def qwestions(str):
    str = str.lower()
    files = get_files_from_text(str)
    print(files)
    for i in files:
        print_img(i)
    
    
    
```
